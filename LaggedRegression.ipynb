```python
import numpy as N
from matplotlib import pyplot as plt

length = 20
m = 1
stdv = .1
maxLag = 5

actualLag = 2

x = np.random.random(length)
y = np.random.random(length)
y[maxLag-actualLag:-actualLag] = m*x[maxLag-actualLag:-actualLag] + np.random.normal(loc=0, scale=stdv, size=length-maxLag)

xL = N.zeros([maxLag+1,length-maxLag])

xL[0,:] = x[maxLag:]
for i in xrange(1,maxLag+1):
    xL[i,:] = x[maxLag-i:-i]
yL = y[maxLag-actualLag:-actualLag]

plt.scatter(x[maxLag-actualLag:-actualLag],y[maxLag-actualLag:-actualLag])
```

```python
plt.scatter(xL[0,:],yL)
plt.scatter(xL[1,:],yL,color='red')
plt.scatter(xL[2,:],yL,color='green')
```

```python
for i in xrange(maxLag):
    plt.scatter(xL[i],yL)
    rFit = np.polyfit(xL[i],yL,1,full = True)        
    print 'residual:',rFit[1][0]
    rBestLine = np.poly1d(rFit[0])
    x_lin = np.linspace(min(xL[i]),max(xL[i]),1000)
    plt.plot(x_lin,rBestLine(x_lin),'k',lw=2)
    plt.show()
```
